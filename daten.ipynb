```python
import geopandas
import matplotlib.pyplot as plt
```

```python
# Deine Bestell-Daten
data = {
    "Country": [
        "Germany", "United States of America", "Brazil", "France", "United Kingdom", "Venezuela", "Austria", "Sweden",
        "Canada", "Italy", "Mexico", "Spain", "Finland", "Ireland", "Belgium",
        "Denmark", "Switzerland", "Argentina", "Portugal", "Poland", "Norway"
    ],
    "Orders": [
        122, 122, 83, 77, 56, 46, 40, 37, 30, 28, 28, 23, 22, 19, 19, 18, 18, 16,
        13, 7, 6
    ]
}
df = pd.DataFrame(data)
# Pfad zum entpackten Shapefile
shapefile_path = r"C:\Users\Admin\Desktop\ne_110m_admin_0_countries\ne_110m_admin_0_countries.shp"
# Weltkarte laden
world = geopandas.read_file(shapefile_path)
# Daten mit den Kartendaten zusammenführen
merged = world.set_index('ADMIN').join(df.set_index('Country'))
# Skalierung der Kreisgrößen festlegen
min_orders = merged['Orders'].min()
max_orders = merged['Orders'].max()
scale_factor = 2000  # Anpassbarer Faktor für die Kreisgröße
# Koordinaten für Paris
paris_lon = 2.3522
paris_lat = 48.8566
# Plot erstellen
fig, ax = plt.subplots(1, 1, figsize=(12, 6))
world.plot(ax=ax, color='lightgray', edgecolor='black')
# Punkte und Text für die Bestellungen hinzufügen
for idx, row in merged.dropna(subset=['Orders']).iterrows():
    if idx == 'France':
        lon = paris_lon
        lat = paris_lat
    else:
        try:
            lon = row['geometry'].centroid.x
            lat = row['geometry'].centroid.y
        except AttributeError:
            print(f"Geometrie für {idx} ist ungültig und wird übersprungen.")
            continue
    order_size = (row['Orders'] - min_orders) / (max_orders - min_orders) * scale_factor + 50
    ax.plot(lon, lat, 'o', markersize=order_size**0.5, color='red', alpha=0.6)
    ax.text(lon, lat, f"{int(row['Orders'])}", fontsize=8, ha='center', va='center', color='black')
# Titel und Achsen entfernen
ax.set_title('Bestellmengen pro Land')
ax.set_axis_off()
# Bild speichern
#plt.savefig('orders_by_country_map_matplotlib_with_numbers.png')
#plt.show()
```
